```python
import matplotlib.pyplot as plt

# car
x = [0, 6, 6, 5, 4, 1, 0, 0]
y = [1, 1, 3, 4, 4, 3, 3, 1]

fig = plt.figure()

ax = fig.add_subplot(111)

ax.plot(x, y)

# wheels
ax.add_patch(plt.Circle((1, 1), 0.5, fill=True, facecolor='black'))
ax.add_patch(plt.Circle((5, 1), 0.5, fill=True, facecolor='black'))

# windows
ax.add_patch(plt.Rectangle((2, 2), 2, 1.5, fill=False))
ax.add_patch(plt.Rectangle((3.5, 2), 1, 1.5, fill=False))

# set the labels for the x-axis and y-axis
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# set the title for the plot
ax.set_title('Car')

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Create a figure and axis
fig, ax = plt.subplots()
# Draw the roof
roof_x = [1, 5, 3, 1]
roof_y = [5, 5, 7, 5]
ax.plot(roof_x, roof_y, 'darkred', linewidth=2) # 'darkred' color forroof
# Draw the walls
wall_x = [1, 1, 5, 5, 1]
wall_y = [1, 5, 5, 1, 1]
ax.plot(wall_x, wall_y, 'saddlebrown', linewidth=2) # 'saddlebrown' colorfor walls
# Draw the windows
window1_x = [2, 3, 3, 2, 2]
window1_y = [2, 2, 3, 3, 2]
ax.plot(window1_x, window1_y, 'lightblue', linewidth=2) # 'lightblue'color for window 1
# Set axis limits and labels
ax.set_xlim([0, 6]) # type: ignore
ax.set_ylim([0, 8]) # type: ignore
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('2D House Scenery Object')
# Display the plot
plt.show()
```
