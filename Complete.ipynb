Bresen Line

```python
import matplotlib.pyplot as plt

def bresenham_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    x = x1
    y = y1
    x_inc = 1 if dx > 0 else -1
    y_inc = 1 if dy > 0 else -1
    dx = abs(dx)
    dy = abs(dy)
    if dx > dy:
        ai = (dy - dx) * 2
        bi = dy * 2
        d = bi - dx
        while x != x2:
            if d >= 0:
                y += y_inc
                d += ai
            else:
                d += bi
            x += x_inc
            yield x, y
    else:
        ai = (dx - dy) * 2
        bi = dx * 2
        d = bi - dy
        while y != y2:
            if d >= 0:
                x += x_inc
                d += ai
            else:
                d += bi
            y += y_inc
            yield x, y

# Taking input from the user
x1 = int(input("Enter the value of x1: "))
y1 = int(input("Enter the value of y1: "))
x2 = int(input("Enter the value of x2: "))
y2 = int(input("Enter the value of y2: "))
for x, y in bresenham_line(x1, y1, x2, y2):
    print(f"({x}, {y})")
# Drawing the line using Bresenham's algorithm
x_values = []
y_values = []
for x, y in bresenham_line(x1, y1, x2, y2):
    x_values.append(x)
    y_values.append(y)

# Plotting the line on a graph
plt.plot(x_values, y_values)
plt.title("Bresenham's Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Mid Circle

```python
import matplotlib.pyplot as plt

def mid_point_circle(center_x, center_y, radius):
    x = 0
    y = radius
    d = 1 - radius
    points = set()
    
    while x <= y:
        points.add((x + center_x, y + center_y))
        points.add((y + center_x, x + center_y))
        points.add((-x + center_x, y + center_y))
        points.add((-y + center_x, x + center_y))
        points.add((-x + center_x, -y + center_y))
        points.add((-y + center_x, -x + center_y))
        points.add((x + center_x, -y + center_y))
        points.add((y + center_x, -x + center_y))
        
        if d < 0:
            d += 2 * x + 3
        else:
            d += 2 * (x - y) + 5
            y -= 1
        
        x += 1
    
    return points

# Taking input from the user
center_x = int(input("Enter the x-coordinate of the center of the circle: "))
center_y = int(input("Enter the y-coordinate of the center of the circle: "))
radius = int(input("Enter the radius of the circle: "))

# Drawing the circle using Midpoint Circle Drawing Algorithm
points = mid_point_circle(center_x, center_y, radius)

# Converting the set of points into separate lists of x and y coordinates
x_values, y_values = zip(*points)

# Plotting the circle on a graph
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.scatter(x_values, y_values, s=10)
plt.title("Midpoint Circle Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Brsen Circle

```python
import matplotlib.pyplot as plt

def bresenham_circle(center_x, center_y, radius):
    x = 0
    y = radius
    d = 3 - 2 * radius
    points = set()
    
    while x <= y:
        points.add((x + center_x, y + center_y))
        points.add((y + center_x, x + center_y))
        points.add((-x + center_x, y + center_y))
        points.add((-y + center_x, x + center_y))
        points.add((-x + center_x, -y + center_y))
        points.add((-y + center_x, -x + center_y))
        points.add((x + center_x, -y + center_y))
        points.add((y + center_x, -x + center_y))
        
        if d < 0:
            d = d + 4 * x + 6
        else:
            d = d + 4 * (x - y) + 10
            y -= 1
        
        x += 1
    
    return points

# Taking input from the user
center_x = int(input("Enter the x-coordinate of the center of the circle: "))
center_y = int(input("Enter the y-coordinate of the center of the circle: "))
radius = int(input("Enter the radius of the circle: "))

# Drawing the circle using Bresenham's Circle Drawing Algorithm
points = bresenham_circle(center_x, center_y, radius)

# Converting the set of points into separate lists of x and y coordinates
x_values, y_values = zip(*points)

# Plotting the circle on a graph
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.scatter(x_values, y_values, s=10)
plt.title("Bresenham's Circle Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Point-Clip

```python

import matplotlib.pyplot as plt


def point_clipping(x, y, x_min, x_max, y_min, y_max):
    if(x_min <= x <= x_max) and (y_min <= y <= y_max):
        plt.plot(x, y, '.r')
        print("Inside the clipping window")
    else:
        plt.plot(x, y, '.r')
        print("Outside the clipping window")


def main():
    x_min = float(input("Enter x_min: "))
    x_max = int(input("Enter x_max: "))
    y_min = int(input("Enter y_min: "))
    y_max = int(input("Enter y_max: "))

    n = int(input("Enter no of points to clip: "))
    plt.plot([x_min, y_min], [x_min, y_max], 'yellow')
    plt.plot([x_min, y_max], [x_max, y_max], 'yellow')
    plt.plot([x_max, y_max], [x_max, y_min], 'yellow')
    plt.plot([x_max, y_min], [x_min, y_min], 'yellow')

    plt.title("Point Clipping Algorithm")
    plt.xlabel("X coordinates")
    plt.ylabel("Y coordinates")

    for i in range(n):
        x = int(input("Enter x coordinate: "))
        y = int(input("Enter y coordinate: "))
        point_clipping(x, y, x_min, x_max, y_min, y_max)

    plt.axis('scaled')
    plt.show()


if __name__ == "__main__":
    main()

```

Cohen-S Line

```python
import matplotlib.pyplot as plt

INSIDE = 0  # 0000
LEFT = 1  # 0001
RIGHT = 2  # 0010
BOTTOM = 4  # 0100
TOP = 8  # 1000


def compute_code(x, y, x_min, x_max, y_min, y_max):
    code = INSIDE
    if x < x_min:
        code |= LEFT
    elif x > x_max:
        code |= RIGHT
    if y < y_min:
        code |= BOTTOM
    elif y > y_max:
        code |= TOP

    return code


def cohen_sutherland(x1, y1, x2, y2, x_min, x_max, y_min, y_max):
    code1 = compute_code(x1, y1, x_min, x_max, y_min, y_max)
    code2 = compute_code(x2, y2, x_min, x_max, y_min, y_max)

    accept = False
    a = [x1, x2]
    b = [y1, y2]
    plt.plot(a, b, color='green')

    while True:

        if code1 == 0 and code2 == 0:
            accept = True
            break

        elif (code1 & code2) != 0:  # AND not 0 => line is completely outside
            break

        else:
            x = 1.0
            y = 1.0
            if code1 != 0:
                code_out = code1
            else:
                code_out = code2

            if code_out & TOP:
                x = x1 + (x2 - x1) * (y_max - y1) / (y2 - y1)
                y = y_max
            elif code_out & BOTTOM:
                x = x1 + (x2 - x1) * (y_min - y1) / (y2 - y1)
                y = y_min
            elif code_out & RIGHT:
                y = y1 + (y2 - y1) * (x_max - x1) / (x2 - x1)
                x = x_max
            elif code_out & LEFT:
                y = y1 + (y2 - y1) * (x_min - x1) / (x2 - x1)
                x = x_min

            if code_out == code1:
                x1 = x
                y1 = y
                code1 = compute_code(x1, y1, x_min, x_max, y_min, y_max)

            else:
                x2 = x
                y2 = y
                code2 = compute_code(x2, y2, x_min, x_max, y_min, y_max)

    X = [x_min, x_max, x_max, x_min, x_min]
    Y = [y_min, y_min, y_max, y_max, y_min]

    plt.title("Cohen Sutherland Line Clipping Algorithm")
    plt.xlabel("X coordinates")
    plt.ylabel("Y coordinates")
    plt.plot(X, Y, color='yellow')
    plt.axis('scaled')

    if accept:
        print("\nLine clipped from (%.2f, %.2f) to (%.2f, %.2f)" % (x1, y1, x2, y2))
        plt.scatter(x1, y1, color="red")
        plt.text(x1, y1, "Point A")
        plt.scatter(x2, y2, color="red")
        plt.text(x2,y2,"PointB")
    else:
        print("Line rejected")

    plt.show()


if __name__ == "__main__":
    x_min = int(input("Enter x_min: "))
    x_max = int(input("Enter x_max: "))
    y_min = int(input("Enter y_min: "))
    y_max = int(input("Enter y_max: "))

    x1, y1 = map(int, input("Enter x1 and y1: ").split())
    x2, y2 = map(int, input("Enter x2 and y2: ").split())

    plt.scatter(x1, y1, color="blue")
    plt.scatter(x2, y2, color="blue")

    cohen_sutherland(x1, y1, x2, y2, x_min, x_max, y_min, y_max)
```

Liang B Line

```python
# Liang-Barsky Line Clipping algorithm
import matplotlib.pyplot as plt

def liangBarsky(x1, y1, x2, y2, xmin, ymin, xmax, ymax):
    # calculate differences
    dx = x2 - x1
    dy = y2 - y1
    # initialize p and q arrays
    p = [-dx, dx, -dy, dy]
    q = [x1 - xmin, xmax - x1, y1 - ymin, ymax - y1]
    # initialize u1 and u2
    u1 = 0
    u2 = 1
    # loop through all edges of the clip window
    for i in range(4):
        # handle parallel lines
        if p[i] == 0:
            if q[i] < 0:
                return None
        else:
            # calculate u value
            u = q[i] / p[i]
            # check u values and update u1 and u2
            if p[i] < 0 and u > u1:
                u1 = u
            elif p[i] > 0 and u < u2:
                u2 = u
            # handle invalid lines
            if u1 > u2:
                return None
            # calculate new coordinates of line
            new_x1 = x1 + u1 * dx
            new_y1 = y1 + u1 * dy
            new_x2 = x1 + u2 * dx
            new_y2 = y1 + u2 * dy

    # plot the valid and clipped lines
    plt.plot([x1, x2], [y1, y2], color='blue', label='Valid Line')
    plt.plot([new_x1, new_x2], [new_y1, new_y2],
             color='red', label='Clipped Line')
    # plot the clip window
    plt.plot([xmin, xmin, xmax, xmax, xmin], [ymin, ymax, ymax, ymin, ymin],
             color='black', label='Clip Window')
    # plot the intersection points if exists
    if u1 > 0:
        plt.scatter(new_x1, new_y1, color='red',
                    label='Intersection Point:({: .2f}, {: .2f})'.format(new_x1, new_y1))
    if u2 < 1:
        plt.scatter(new_x2, new_y2, color='red',
                    label='Intersection Point:({: .2f}, {: .2f})'.format(new_x2, new_y2))
    # add labels and title to the plot
    plt.legend()
    plt.title('Liang-Barsky Clipping Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    # show the plot
    plt.show()
    # return the new coordinates of the line
    return (new_x1, new_y1, new_x2, new_y2)


def main():
    # get user input for the clip window and line
    xMin = int(input("Enter Coordinate xMin : "))
    yMin = int(input("Enter Coordinate yMin : "))
    xMax = int(input("Enter Coordinate xMax : "))
    yMax = int(input("Enter Coordinate yMax : "))
    x1 = float(input("Enter Coordinate  x1 : "))
    y1 = float(input("Enter Coordinate  y1 : "))
    x2 = float(input("Enter Coordinate  x2 : "))
    y2 = float(input("Enter Coordinate  y2 : "))
    liangBarsky(x1, y1, x2, y2, xMin, yMin, xMax, yMax)
    plt.show()

if __name__ == "__main__":
    main()
```

S-Hodgman Polygon

```python
import matplotlib.pyplot as plt

# function to check if a point is inside or outside the clip edge
def inside(p, cp1, cp2):
    return (cp2[0]-cp1[0])*(p[1]-cp1[1]) > (cp2[1]-cp1[1])*(p[0]-cp1[0])

# function to compute intersection point of two edges
def computeIntersection(cp1, cp2, s, e):
    dc = [cp1[0] - cp2[0], cp1[1] - cp2[1]]
    dp = [s[0] - e[0], s[1] - e[1]]
    n1 = cp1[0] * cp2[1] - cp1[1] * cp2[0]
    n2 = s[0] * e[1] - s[1] * e[0]
    n3 = 1.0 / (dc[0] * dp[1] - dc[1] * dp[0])
    # using determinants to compute intersection point
    return [(n1*dp[0] - n2*dc[0]) * n3, (n1*dp[1] - n2*dc[1]) * n3]

# main function for Sutherland-Hodgman algorithm
def sutherlandHodgman(subject_polygon, clip_polygon):
    output_list = subject_polygon
    cp1 = clip_polygon[-1]
    for clip_vertex in clip_polygon:
        cp2 = clip_vertex
        input_list = output_list
        output_list = []
        s = input_list[-1]
        for subject_vertex in input_list:
            e = subject_vertex
            if inside(e, cp1, cp2):
                if not inside(s, cp1, cp2):
                    # if the edge intersects the clip edge, compute intersection point
                    output_list.append(computeIntersection(cp1, cp2, s, e))
                output_list.append(e)
            elif inside(s, cp1, cp2):
                # if the edge exits the clip edge, compute intersection point
                output_list.append(computeIntersection(cp1, cp2, s, e))
            s = e
        cp1 = cp2
    return output_list

# getting input from user
x_min = int(input("Enter xMin: "))
y_min = int(input("Enter yMin: "))
x_max = int(input("Enter xMax: "))
y_max = int(input("Enter yMax: "))

# defining the clip polygon as a rectangle
clip_polygon = [[x_min, y_min], [x_max, y_min], [x_max, y_max], [x_min, y_max]]

n = int(input("Enter the number of points: "))
subject_polygon = []
for i in range(n):
    x, y = map(int, input(f"Enter Point {i + 1}: ").split())
    subject_polygon.append([x, y])
subject_polygon.append(subject_polygon[0])

# applying Sutherland-Hodgman algorithm to clip the subject polygon
clipped_polygon = sutherlandHodgman(subject_polygon, clip_polygon)

# plotting the original clipping rectangle, the original subject polygon, and the clipped polygon
x = [x_min, x_max, x_max, x_min, x_min]
y = [y_min, y_min, y_max, y_max, y_min]

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(x, y, 'r--', label='Clipping Window')

ax.plot(*zip(*subject_polygon), 'bo-', label='Subject Polygon')
ax.plot(*zip(*clipped_polygon), 'k-', label='Clipped Polygon')
ax.legend()
plt.show()
```

2D scaling

```python
import matplotlib.pyplot as plt
import numpy as np

triangle_points = np.array([[50, 100], [50, 200], [100, 100]])
x_scaling_factor = 2
y_scaling_factor = 2
scaling_matrix = np.array([[x_scaling_factor, 0], [0, y_scaling_factor]])
scaled_triangle_points = triangle_points.dot(scaling_matrix)
triangle_x = triangle_points[:, 0]
triangle_y = triangle_points[:, 1]
scaled_triangle_x = scaled_triangle_points[:, 0]
scaled_triangle_y = scaled_triangle_points[:, 1]
# Plotting original and scaled triangles
plt.figure(figsize=(8, 6))
plt.plot(triangle_x, triangle_y, 'r', label='Original Triangle')
plt.plot(scaled_triangle_x, scaled_triangle_y, 'g', label='Scaled Triangle')
plt.fill(triangle_x, triangle_y, 'r', alpha=0.3, label='Original Triangle Fill')
plt.fill(scaled_triangle_x, scaled_triangle_y, 'g', alpha=0.3,
label='Scaled Triangle Fill')
plt.plot(np.append(triangle_x, triangle_x[0]), np.append(triangle_y,
triangle_y[0]), 'r', label='Triangle Sides')
plt.plot(np.append(scaled_triangle_x, scaled_triangle_x[0]),
np.append(scaled_triangle_y, scaled_triangle_y[0]), 'g', label='Scaled Triangle Sides')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Scaling Transformation')
plt.legend()
plt.grid()
plt.show()
```

2D Shearing

```python
import matplotlib.pyplot as plt

# Input number of sides of the polygon
n = int(input("Enter the number of sides of the polygon: "))

# Input coordinates of the polygon vertices
vertices = []
for i in range(n):
    x, y = map(int, input(
        "Enter the x, y coordinates of vertex {}: ".format(i+1)).split())
    vertices.append([x, y])

# Append the first vertex again to the end to complete the polygon
vertices.append(vertices[0])

# Extract the x and y coordinates separately
x_coords = [v[0] for v in vertices]
y_coords = [v[1] for v in vertices]

# Plot the original polygon and its coordinates
plt.plot(x_coords, y_coords, label="Original")
for i, vertex in enumerate(vertices[:-1]):
    plt.text(vertex[0], vertex[1], f"({vertex[0]},{vertex[1]})", fontsize=8)

shear_x = float(input("Enter the shear factor (in x direction): "))
shear_y = float(input("Enter the shear factor (in y direction): "))
sheared_vertices = [[x + y*shear_x, y + x*shear_y] for [x, y] in vertices]

# Extract the x and y coordinates separately
sheared_x_coords = [v[0] for v in sheared_vertices]
sheared_y_coords = [v[1] for v in sheared_vertices]

plt.plot(sheared_x_coords, sheared_y_coords, label="Sheared")
for i, vertex in enumerate(sheared_vertices[:-1]):
    plt.text(vertex[0], vertex[1],
             f"({vertex[0]:.2f},{vertex[1]:.2f})", fontsize=8)

# Set the aspect ratio to 'equal' for a square plot
plt.axis('equal')

plt.legend()
plt.title("Polygon and its Sheared Version")

plt.show()
```

2D Rotation

```python
# rotation
import matplotlib.pyplot as plt
import math

n = int(input("Enter the number of sides of the polygon: "))

vertices = []
for i in range(n):
    x, y = map(int, input(
        "Enter the x, y coordinates of vertex {}: ".format(i+1)).split())
    vertices.append([x, y])

# Append the first vertex again to the end to complete the polygon
vertices.append(vertices[0])

# Extract the x and y coordinates separately
x_coords = [v[0] for v in vertices]
y_coords = [v[1] for v in vertices]

# Plot the original polygon and its coordinates
plt.plot(x_coords, y_coords, label="Original")
for i, vertex in enumerate(vertices[:-1]):
    plt.text(vertex[0], vertex[1], f"({vertex[0]},{vertex[1]})", fontsize=8)

angle = float(input("Enter the angle of rotation (in degrees): "))
theta = math.radians(angle)
rotated_vertices = [[x*math.cos(theta) - y*math.sin(theta),
                     x*math.sin(theta) + y*math.cos(theta)] for [x, y] in vertices]

rotated_x_coords = [v[0] for v in rotated_vertices]
rotated_y_coords = [v[1] for v in rotated_vertices]

plt.plot(rotated_x_coords, rotated_y_coords, label="Rotated")
for i, vertex in enumerate(rotated_vertices[:-1]):
    plt.text(vertex[0], vertex[1],
             f"({vertex[0]:.2f},{vertex[1]:.2f})", fontsize=8)

plt.legend()
plt.title("Polygon and its Rotated Version")

plt.show()
```
